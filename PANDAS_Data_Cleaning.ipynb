```python
# importing library 
import pandas as pd 

# importing dataset
df = pd.read_excel(r'C:\Users\ARJUN\Python_projects\Inconsistent_Data.xlsx')
df
```

```python
# dropping duplicates

df = df.drop_duplicates()
df
```

```python
# dropping unnecessary columns

df = df.drop(columns = ['Not_Useful_Column'])
df
```

```python
# removing leading and trailing characters 

df['Last_Name'] = df['Last_Name'].str.strip('/._')
df
```

```python
# converting the 'Phone_Number' column into a string for data cleansing

df['Phone_Number'] = df['Phone_Number'].astype(str)
df['Phone_Number'] = df['Phone_Number'].str.replace('[^0-9]', '', regex = True)
df
```

```python
# formatting the 'Phone_Number' column

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:9])
df
```

```python
# removing double hyphens ('--')

df['Phone_Number'] = df['Phone_Number'].str.replace('--', '')
df
```

```python
# splitting the address column

df[['Street Address', 'State', 'Zip Code']] = df['Address'].str.split(',', n = 2, expand = True) # expand = True converts the result split into a DataFrame, n = 2 equals number of splits
df
```

```python
# standardizing the 'Paying Customer' column

df['Paying Customer'] = df['Paying Customer'].str.replace('^Y$', 'Yes', regex = True)
df['Paying Customer'] = df['Paying Customer'].str.replace('^N$', 'No', regex = True)
df
```

```python
# standardizing the 'Do_Not_Contact' column

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('^Y$', 'Yes', regex = True)
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('^N$', 'No', regex = True)
df
```

```python
# replacing 'N/a' in columns with ''

df.replace('N/a', '', inplace = True)
df
```

```python
# populating NaN values with ''

df.fillna('', inplace = True)
df
```

```python
# removing rows where 'Do_Not_Contact' column has the value 'Yes'

for i in df.index:
    if df.loc[i, 'Do_Not_Contact'] == 'Yes':
        df.drop(i, inplace = True)

df
```

```python
# populating blank rows in 'Do_Not_Contact' with 'No'

df['Do_Not_Contact'] = df['Do_Not_Contact'].apply(lambda x: x.strip() + 'No' if x.strip() == '' else x) # if the stripped value is empty then, concatenate 'No' to the stripped value

df
```

```python
# dropping rows where 'Phone_Number' is an empty string

for i in df.index:
    if df.loc[i, 'Phone_Number'] == '':
        df.drop(i, inplace = True)
        
df
```

```python
# resetting the index

df.reset_index(drop = True, inplace = True) # drop = True discards the old index 
```

```python
# dropping the 'Zip Code' column

df = df.drop(columns = ['Zip Code'])
df
```

# Summary

1. **Removed Special Characters:** Cleaned the Last_Name column by stripping unwanted characters like **/**, **.**, and **_**.

2. **Cleaned Phone_Number Formatting:** Removed non-numeric characters from phone numbers and reformatted them to a standard XXX-XXX-XXX format.

3. **Split Address Column:** Separated the Address column into Street Address, State, and Zip Code based on commas.

4. **Standardized Paying Customer Values:** Replaced occurrences of 'Y' with 'Yes' and 'N' with 'No' in the Paying Customer column.

5. **Removed Missing Values:** Dropped rows with missing phone numbers to ensure complete data.

6. **Dropped Unnecessary Columns:** Removed irrelevant columns to reduce clutter and improve dataset focus.

7. **Reset Index:** Reset the DataFrame index after data cleaning operations to ensure sequential numbering.

8. **Dropped Rows Based on Condition:** Removed rows where individuals had opted for "Do Not Contact."

9. **Handled Empty Values in Do_Not_Contact:** Filled empty or blank values in the Do_Not_Contact column with 'No'.
